# 1. Preparación de los datos

```python
import pandas as pd
```

```python
df = pd.read_csv('./dataset/CovidData.csv')
print(len(df))
df.head()
```

```python
df.dtypes
```

```python
clasiffication_final = pd.to_numeric(df.CLASIFFICATION_FINAL, errors='coerce')


df.CLASIFFICATION_FINAL = pd.to_numeric(df.CLASIFFICATION_FINAL, errors='coerce')


df.CLASIFFICATION_FINAL = df.CLASIFFICATION_FINAL.fillna(0)


df[clasiffication_final.notnull()][['USMER', 'CLASIFFICATION_FINAL']]
```

```python
replacer = lambda str: str.lower().str.replace(' ', '_')
df.columns = replacer(df.columns.str)
for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacer(df[col].str)
df.head()
```

```python
numerical = ['medical_unit', 'sex', 'patient_type', 'intubed', 'pneumonia', 'age', 'pregnant', 'diabetes', 'copd', 'asthma',
             'inmsupr', 'hipertension', 'other_disease', 'cardiovascular', 'obesity', 'renal_chronic', 'tobacco', 'clasiffication_final']
categorical = ['date_died']
df[numerical].nunique()
```

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)

df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=1)
y_train = df_train.icu.values
y_val = df_val.icu.values

del df_train['icu']
del df_val['icu']

df_train.head().T
```

# 2. Análisis de importancia de propiedades

```python
global_mean = df_train_full.icu.mean()
round(global_mean, 3)
```

```python
usmer1_mean = df_train_full[df_train_full.usmer == 1].icu.mean()
print(round(usmer1_mean, 3))
usmer2_mean = df_train_full[df_train_full.usmer == 2].icu.mean()
print(round(usmer2_mean, 3))
```

```python
patient1 = df_train_full[df_train_full.patient_type == 1].icu.mean()
print(round(patient1, 3))
patient2 = df_train_full[df_train_full.patient_type == 2].icu.mean()
print(round(patient2, 3))
```

```python
from sklearn.metrics import mutual_info_score

calculate_mi = lambda col: mutual_info_score(col, df_train_full.icu)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='mi')
df_mi
```

```python
print(df_train_full[numerical].corrwith(df_train_full.icu))

print(round(df_train_full[df_train_full.medical_unit <= 2].icu.mean(), 3))
print(round(df_train_full[(df_train_full.medical_unit > 3) & 
                          (df_train_full.medical_unit <= 12)].icu.mean(), 3))
print(round(df_train_full[df_train_full.medical_unit > 12].icu.mean(), 3))

print(round(df_train_full[df_train_full.age < 20].icu.mean(), 3))
print(round(df_train_full[(df_train_full.age > 21) & 
                          (df_train_full.age <= 50)].icu.mean(), 3))
print(round(df_train_full[df_train_full.age > 50].icu.mean(), 3))
```

# 3. Ingeniería de propiedades

```python
train_dict = df_train[categorical + numerical].to_dict(orient='records')
dict(sorted(train_dict[0].items()))
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
X_train[0]
```

```python
dv.get_feature_names_out()
```

# 4. Entrenamiento del modelo

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
y_pred = model.predict_proba(X_val)
y_pred
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
churn = y_pred >= 0.5
churn
```

```python
round((y_val == churn).mean(), 3) 
```

# 5. Serialización del modelo

```python
import pickle
import joblib
```

```python
with open('./models/ICU-model.pck', 'wb') as f:
    pickle.dump((dv, model), f)

with open('./models/ICU-model_joblib.pck', 'wb') as f:
    joblib.dump((dv, model), f)
```

```python
with open('./models/ICU-model.pck', 'rb') as f:
    dv, model = pickle.load(f)
    X_val = dv.transform(val_dict)
    y_pred = model.predict_proba(X_val)

print(y_pred)

with open('./models/ICU-model_joblib.pck', 'rb') as f:
    dv, model = joblib.load(f)
    X_val = dv.transform(val_dict)
    y_pred = model.predict_proba(X_val)

print(y_pred)
```
